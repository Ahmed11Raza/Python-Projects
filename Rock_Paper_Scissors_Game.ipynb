#1st Method:

```python
import random

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "user"
    else:
        return "computer"

def display_score(user_score, computer_score):
    print(f"\nScores:\nYou: {user_score}\nComputer: {computer_score}\n")

def play_game():
    user_score = 0
    computer_score = 0
    rounds = 0

    while True:
        print("\n--- Round", rounds + 1, "---")
        user_choice = input("Enter your choice (rock, paper, scissors or 'q' to quit): ").lower()

        if user_choice == 'q':
            print("Game Over!")
            display_score(user_score, computer_score)
            break

        if user_choice not in ["rock", "paper", "scissors"]:
            print("Invalid choice. Please try again.")
            continue

        computer_choice = get_computer_choice()
        print(f"Computer chose: {computer_choice}")

        winner = determine_winner(user_choice, computer_choice)
        if winner == "user":
            print("You win this round!")
            user_score += 1
        elif winner == "computer":
            print("Computer wins this round!")
            computer_score += 1
        else:
            print("It's a tie this round!")

        rounds += 1
        display_score(user_score, computer_score)

if __name__ == "__main__":
    play_game()
```

#2nd method:

```python
import random

CHOICES = ['rock', 'paper', 'scissors']
MAX_ROUNDS = 5

def get_computer_choice():
    return random.choice(CHOICES)

def determine_winner(player_choice, computer_choice):
    if player_choice == computer_choice:
        return 'tie'
    win_conditions = {
        'rock': 'scissors',
        'paper': 'rock',
        'scissors': 'paper'
    }
    return 'player' if win_conditions[player_choice] == computer_choice else 'computer'

def display_score(player_score, computer_score):
    print(f"\nScore - You: {player_score}, Computer: {computer_score}")

def display_final_result(player_score, computer_score):
    print("\nGame Over!")
    print(f"Final Score - You: {player_score}, Computer: {computer_score}")
    if player_score > computer_score:
        print("Congratulations! You win the game!")
    elif player_score < computer_score:
        print("The computer wins the game. Better luck next time!")
    else:
        print("The game ends in a tie!")

def get_player_choice():
    while True:
        choice = input('Enter your choice (rock/paper/scissors): ').lower().strip()
        if choice in CHOICES:
            return choice
        print("Invalid choice. Please choose rock, paper, or scissors.")

def play_round(round_number):
    print(f"\nRound {round_number}")
    player_choice = get_player_choice()
    computer_choice = get_computer_choice()
    print(f"Computer chose: {computer_choice}")

    result = determine_winner(player_choice, computer_choice)
    if result == 'tie':
        print("It's a tie!")
    elif result == 'player':
        print("You win this round!")
    else:
        print("Computer wins this round!")

    return result

def play_game():
    player_score = 0
    computer_score = 0

    print("Welcome to Rock, Paper, Scissors!")
    print(f"Best of {MAX_ROUNDS} rounds. May the best player win!\n")

    for round in range(1, MAX_ROUNDS + 1):
        result = play_round(round)
        if result == 'player':
            player_score += 1
        elif result == 'computer':
            computer_score += 1

        display_score(player_score, computer_score)

        if round < MAX_ROUNDS:
            play_next = input('\nPress Enter to continue to the next round, or type "quit" to end the game: ').lower().strip()
            if play_next == 'quit':
                break

    display_final_result(player_score, computer_score)

if __name__ == "__main__":
    play_game()
```
