#  Flight Reservation System:

```python
class FlightReservationSystem:
    def __init__(self):
        self.flights = {
            "101": {"destination": "New York", "price": 500, "seats": 20, "airline": "Delta Airlines"},
            "102": {"destination": "Toronto", "price": 550, "seats": 18, "airline": "Air Canada"},
            "202": {"destination": "London", "price": 700, "seats": 15, "airline": "British Airways"},
            "303": {"destination": "Dubai", "price": 400, "seats": 25, "airline": "Emirates"},
            "404": {"destination": "Tokyo", "price": 800, "seats": 10, "airline": "Japan Airlines"},
            "505": {"destination": "Paris", "price": 600, "seats": 12, "airline": "Air France"},
            "606": {"destination": "Sydney", "price": 900, "seats": 8, "airline": "Qantas"},
            "707": {"destination": "Beijing", "price": 750, "seats": 14, "airline": "Air China"},
            "808": {"destination": "Berlin", "price": 650, "seats": 11, "airline": "Lufthansa"},
            "909": {"destination": "Cape Town", "price": 850, "seats": 9, "airline": "South African Airways"},
            "1001": {"destination": "Bangkok", "price": 700, "seats": 13, "airline": "Thai Airways"},
            "1101": {"destination": "Singapore", "price": 720, "seats": 15, "airline": "Singapore Airlines"},
            "1201": {"destination": "Moscow", "price": 670, "seats": 12, "airline": "Aeroflot"},
            "1301": {"destination": "Mumbai", "price": 550, "seats": 16, "airline": "Air India"},
            "1401": {"destination": "Jeddah", "price": 500, "seats": 20, "airline": "Flynas"},
            "1501": {"destination": "Los Angeles", "price": 850, "seats": 10, "airline": "United Airlines"},
            "1601": {"destination": "Rome", "price": 600, "seats": 14, "airline": "Alitalia"},
        }
        self.bookings = []

    def display_flights(self):
        print("\nAvailable Flights:")
        print("Flight ID | Destination       | Price | Seats Available | Airline")
        for flight_id, details in self.flights.items():
            print(f"{flight_id}      | {details['destination']} | ${details['price']} | {details['seats']} | {details['airline']}")

    def book_ticket(self):
        self.display_flights()
        flight_id = input("\nEnter the Flight ID you want to book: ")

        if flight_id not in self.flights:
            print("Invalid Flight ID. Please try again.")
            return

        flight = self.flights[flight_id]

        if flight['seats'] == 0:
            print("Sorry, no seats available on this flight.")
            return

        name = input("Enter your name: ")
        num_tickets = int(input("Enter the number of tickets to book: "))

        if num_tickets > flight['seats']:
            print(f"Only {flight['seats']} seats are available. Please reduce the number of tickets.")
            return

        total_cost = num_tickets * flight['price']
        print(f"The total cost is ${total_cost}.")
        confirm = input("Confirm booking? (yes/no): ").strip().lower()

        if confirm == "yes":
            flight['seats'] -= num_tickets
            self.bookings.append({"name": name, "flight_id": flight_id, "tickets": num_tickets, "cost": total_cost})
            print(f"Booking confirmed for {name}! {num_tickets} ticket(s) booked on flight {flight_id}.")
        else:
            print("Booking cancelled.")

    def view_bookings(self):
        print("\nYour Bookings:")
        if not self.bookings:
            print("No bookings found.")
            return

        for booking in self.bookings:
            print(f"Name: {booking['name']}, Flight ID: {booking['flight_id']}, Tickets: {booking['tickets']}, Total Cost: ${booking['cost']}")

    def run(self):
        while True:
            print("\nWelcome to the Flight Reservation System")
            print("1. View Available Flights")
            print("2. Book a Ticket")
            print("3. View Your Bookings")
            print("4. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.display_flights()
            elif choice == "2":
                self.book_ticket()
            elif choice == "3":
                self.view_bookings()
            elif choice == "4":
                print("Thank you for using the Flight Reservation System. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")


if __name__ == "__main__":
    system = FlightReservationSystem()
    system.run()
```
